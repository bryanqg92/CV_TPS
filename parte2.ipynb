```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Leer las imágenes en escala de grises
img1 = cv2.imread('TP1\Material_TPs\TP1\img1_tp.png', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('TP1\Material_TPs\TP1\img2_tp.png', cv2.IMREAD_GRAYSCALE)

# Mostrar las imágenes
plt.subplot(1,2,1),plt.imshow(img1,cmap = 'gray')
plt.title('img1_tp.png'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(img2,cmap = 'gray')
plt.title('img2_tp.png'), plt.xticks([]), plt.yticks([])
plt.show()
```

# Observaciones de comparación de histogramas
1. A pesar de que el contenido de las imágenes son muy diferentes una de la otras, los histogramas evidencian que su distribución es muy similar por no decir que casi igual. lo cual puede indicarnos algunas características generales de la imagen.
2. Como se menciona en el punto anterior puede brindar características generales de la imagen, como iluminación o uniformidad en las regiones lo cual por si sola no puede ser relevante en un modelo de clasificación o detección, considero que puede funcionar como feature pero acompañada de otras características.

```python
fig = plt.figure()
hist1, bins1 = np.histogram(img1.ravel(), 300, [0, 256])
ax3 = plt.subplot(211)
ax3.plot(hist1)
ax3.set_title("Histograma de img1_tp.png")

hist2, bins2 = np.histogram(img2.ravel(), 300, [0, 256])
ax4 = plt.subplot(212)
ax4.plot(hist2)
ax4.set_title("Histograma de img2_tp.png")

plt.subplots_adjust(hspace=0.5)
plt.show()
```
