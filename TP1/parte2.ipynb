```python
#%matplotlib
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Leer las im√°genes en escala de grises
img1 = cv2.imread('C:/Users/leand/Documents/LEANDRO/UBA/CEIA/III/CV/CV_TPS/TP1/Material_TPs/TP1/img1_tp.png', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('C:/Users/leand/Documents/LEANDRO/UBA/CEIA/III/CV/CV_TPS/TP1/Material_TPs/TP1/img2_tp.png', cv2.IMREAD_GRAYSCALE)
```

```python
# Mostrar las im√°genes
plt.subplot(1,2,1),plt.imshow(img1,cmap = 'gray')
plt.title('img1_tp.png'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(img2,cmap = 'gray')
plt.title('img2_tp.png'), plt.xticks([]), plt.yticks([])
plt.show(block=False)
```

# Observaciones de comparaci√≥n de histogramas
1. A pesar de que el contenido de las im√°genes son muy diferentes una de la otra, los histogramas evidencian que su distribuci√≥n es muy similar por no decir que casi igual. lo cual puede indicarnos algunas caracter√≠sticas generales de la imagen como iluminaci√≥n o uniformidad en las regiones que por si sola no puede ser relevante en un modelo de clasificaci√≥n o detecci√≥n, considero que puede funcionar como feature pero acompa√±ada de otras caracter√≠sticas.

```python
fig = plt.figure()
hist1, bins1 = np.histogram(img1.ravel(), 300, [0, 256])
ax3 = plt.subplot(211)
ax3.plot(hist1)
ax3.set_title("Histograma de img1_tp.png")

hist2, bins2 = np.histogram(img2.ravel(), 300, [0, 256])
ax4 = plt.subplot(212)
ax4.plot(hist2)
ax4.set_title("Histograma de img2_tp.png")

plt.subplots_adjust(hspace=0.5)
plt.show(block=False)
```
